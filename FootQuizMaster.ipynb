```python
import random
import tkinter as tk
from tkinter import messagebox
from tkinter import simpledialog

class FootballQuizGame:
    def __init__(self, language='fr'):
        self.language = language
        self.initialize_questions()

    def initialize_questions(self):
        if self.language == 'fr':
            # Questions and Answers on the Premier League in French
            self.questions_premier_league = [
                "Quel joueur a réalisé le triplé le plus rapide de la Premier League ?",
                "Quel joueur, avec 653 matchs, a fait le plus d'apparitions en Premier League ?",
                "Trois joueurs partagent le record du plus grand nombre de cartons rouges en Premier League (8). Qui sont-ils ?",
                "Avec 260 buts, qui est le meilleur buteur de l'histoire de la Premier League ?",
                "Quand a eu lieu la saison inaugurale de la Premier League ?",
                "Quelle équipe a remporté le premier titre de Premier League ?",
                "Avec 202 matches sans encaisser de but, quel gardien de but possède le meilleur bilan de la Premier League ?",
                "Combien de clubs ont participé à la saison inaugurale de la Premier League ?",
                "Quels sont les trois joueurs qui se sont partagé le Soulier d'or de la Premier League en 2018-19 ?",
                "Le but le plus rapide de l'histoire de la Premier League a été marqué en 7,69 secondes. Qui l'a marqué ?",
                "Quel joueur détient le record du plus grand nombre de passes décisives délivrées au cours d'une seule saison de la Premier League et combien en a-t-il délivrées ?",
            ]

            self.answers_premier_league = [
                "Sadio Mane (2 minutes 56 secondes pour Southampton contre Aston Villa en 2015).",
                "Gareth Barry",
                "Patrick Vieira, Richard Dunne et Duncan Ferguson",
                "Alan Shearer",
                "1992-93",
                "Manchester United",
                "Petr Cech",
                "22",
                "Pierre-Emerick Aubameyang, Mohamed Salah et Sadio Mane",
                "Shane Long (pour Southampton contre Watford en 2018-19).",
                "Kevin De Bruyne avec 20 passes décisives lors de la saison 2019-2020.",
            ]

            # Questions and Answers on the Coupe du Monde in French
            self.questions_coupe_du_monde = [
                "Il y a eu deux trophées de la Coupe du monde. Quel était le nom du premier ?",
                "Quel pays a remporté la toute première Coupe du monde en 1930 ?",
                "Quel pays a remporté le plus grand nombre de Coupes du monde ?",
                "Trois pays ont remporté deux fois la Coupe du monde. Pouvez-vous les nommer ?",
                "Quel pays a participé à trois finales de la Coupe du monde, mais n'a jamais remporté la compétition ?",
                "La Coupe du monde 2026 sera organisée dans trois pays différents. Pouvez-vous les nommer ?",
                "Dans quelle Coupe du monde Diego Maradona a-t-il marqué son fameux but de la 'main de Dieu' ?",
                "Le nombre record de buts en Coupe du monde est de 16, marqués par qui ?",
                "Trois personnes ont remporté la Coupe du monde en tant que joueur et en tant qu'entraîneur. Mario Zagallo, Didier Deschamps et... pouvez-vous nommer la troisième ?",
                "Deux joueurs anglais ont remporté le Soulier d'or de la Coupe du monde. Qui sont-ils ?",
            ]

            self.answers_coupe_du_monde = [
                "Trophée Jules Rimet",
                "Uruguay",
                "Brésil",
                "Argentine, France et Uruguay",
                "Pays-Bas",
                "États-Unis, Canada et Mexique",
                "Mexique 1986",
                "Miroslav Klose",
                "Franz Beckenbauer",
                "Gary Lineker (1986) et Harry Kane (2018)",
            ]
        
        elif self.language == 'en':
            # Questions and Answers on the Premier League in English
            self.questions_premier_league = [
                "Which player scored the fastest hat-trick in Premier League history?",
                "Which player, with 653 appearances, holds the record for the most Premier League appearances?",
                "Three players share the record for the most red cards in Premier League history (8). Who are they?",
                "With 260 goals, who is the all-time top scorer in Premier League history?",
                "When did the inaugural season of the Premier League take place?",
                "Which team won the first Premier League title?",
                "With 202 clean sheets, which goalkeeper has the best record in Premier League history?",
                "How many clubs participated in the inaugural season of the Premier League?",
                "Who were the three players who shared the Golden Boot in the 2018-19 Premier League season?",
                "The fastest goal in Premier League history was scored in 7.69 seconds. Who scored it?",
                "Which player holds the record for the most assists in a single Premier League season, and how many did he provide?",
            ]

            self.answers_premier_league = [
                "Sadio Mane (2 minutes 56 seconds for Southampton against Aston Villa in 2015).",
                "Gareth Barry",
                "Patrick Vieira, Richard Dunne, and Duncan Ferguson",
                "Alan Shearer",
                "1992-93",
                "Manchester United",
                "Petr Cech",
                "22",
                "Pierre-Emerick Aubameyang, Mohamed Salah, and Sadio Mane",
                "Shane Long (for Southampton against Watford in 2018-19).",
                "Kevin De Bruyne with 20 assists during the 2019-2020 season.",
            ]

            # Questions and Answers on the World Cup in English
            self.questions_coupe_du_monde = [
                "There have been two trophies for the World Cup. What was the name of the first one?",
                "Which country won the very first World Cup in 1930?",
                "Which country has won the most World Cup titles?",
                "Three countries have won the World Cup twice. Can you name them?",
                "Which country has reached three World Cup finals but has never won the competition?",
                "The 2026 World Cup will be hosted in three different countries. Can you name them?",
                "In which World Cup did Diego Maradona score his famous 'Hand of God' goal?",
                "The record for the most goals in a World Cup is 16, scored by whom?",
                "Three individuals have won the World Cup as both a player and a coach. Mario Zagallo, Didier Deschamps, and... can you name the third?",
                "Two English players have won the Golden Boot in the World Cup. Who are they?",
            ]

            self.answers_coupe_du_monde = [
                "Jules Rimet Trophy",
                "Uruguay",
                "Brazil",
                "Argentina, France, and Uruguay",
                "Netherlands",
                "United States, Canada, and Mexico",
                "Mexico 1986",
                "Miroslav Klose",
                "Franz Beckenbauer",
                "Gary Lineker (1986) and Harry Kane (2018)",
            ]




        elif self.language == 'ar':
            # Questions and Answers in Arabic for the Premier League
            self.questions_premier_league = [
                "من هو اللاعب الذي سجل أسرع هاتريك في تاريخ الدوري الإنجليزي الممتاز؟",
                "أي لاعب، بـ 653 مباراة، يحمل الرقم القياسي لأكبر عدد من المشاركات في الدوري الإنجليزي الممتاز؟",
                "ثلاثة لاعبين يشتركون في الرقم القياسي لأكبر عدد من البطاقات الحمراء في تاريخ الدوري الإنجليزي الممتاز (8). من هم؟",
                "مع 260 هدفًا، من هو هداف التاريخ في الدوري الإنجليزي الممتاز؟",
                "متى أقيم موسم الدوري الإنجليزي الممتاز الinaugural؟",
                "أي فريق فاز بلقب الدوري الإنجليزي الممتاز للمرة الأولى؟",
                "مع 202 مباراة بدون استقبال هدف، أي حارس مرمى يحمل أفضل سجل في تاريخ الدوري الإنجليزي الممتاز؟",
                "كم ناديًا شاركوا في موسم الدوري الإنجليزي الممتاز الinaugural؟",
                "من هم الثلاثة لاعبين الذين شاركوا في نيل جائزة الحذاء الذهبي في موسم 2018-19 في الدوري الإنجليزي الممتاز؟",
                "أسرع هدف في تاريخ الدوري الإنجليزي الممتاز سجل في 7.69 ثانية. من سجله؟",
                "أي لاعب يحمل الرقم القياسي لأكبر عدد من التمريرات الحاسمة خلال موسم واحد في الدوري الإنجليزي الممتاز وكم عددها؟",
            ]

            self.answers_premier_league = [
                "ساديو ماني (2 دقيقة و 56 ثانية لصالح ساوثهامبتون ضد أستون فيلا في عام 2015).",
                "غاريث باري",
                "باتريك فييرا، ريتشارد دان ودانكان فيرغيسون",
                "آلان شيرر",
                "1992-93",
                "مانشستر يونايتد",
                "بيتر تشيك",
                "22",
                "بيير إيمريك أوباميانج، محمد صلاح وساديو ماني",
                "شاين لونغ (لصالح ساوثهامبتون ضد واتفورد في موسم 2018-19).",
                "كيفين دي بروين مع 20 تمريرة حاسمة خلال موسم 2019-2020.",
            ]

            # Questions and Answers in Arabic for the World Cup
            self.questions_coupe_du_monde = [
                "هناك كأسين لكأس العالم. ما اسم الأول؟",
                "أي بلد فاز بكأس العالم الأولى في عام 1930؟",
                "أي بلد فاز بأكبر عدد من ألقاب كأس العالم؟",
                "ثلاث دول فازت بكأس العالم مرتين. هل يمكنك تسميتها؟",
                "أي بلد وصل إلى ثلاث نهائيات لكأس العالم ولكنه لم يفز بالبطولة؟",
                "ستستضيف كأس العالم 2026 في ثلاث دول مختلفة. هل يمكنك تسمية هذه الدول؟",
                "في أي كأس العالم سجل دييغو مارادونا هدفه الشهير بـ 'يد الله'؟",
                "الرقم القياسي لأكبر عدد من الأهداف في كأس العالم هو 16، سجله من؟",
                "ثلاثة أفراد فازوا بكأس العالم كلاعبين ومدربين. ماريو زاجالو، ديدييه ديشامب و... هل يمكنك تسمية الثالث؟",
                "لاعبان إنجليزيان فازا بحذاء الذهب في كأس العالم. من هما؟",
            ]

            self.answers_coupe_du_monde = [
                "كأس جول ريميت",
                "أوروغواي",
                "البرازيل",
                "الأرجنتين وفرنسا وأوروغواي",
                "هولندا",
                "الولايات المتحدة وكندا والمكسيك",
                "المكسيك 1986",
                "ميروسلاف كلوزه",
                "فرانز بيكنباور",
                "غاري لاينكر (1986) وهاري كين (2018)",
            ]

        self.questions_and_answers = list(zip(self.questions_premier_league, self.answers_premier_league))
        random.shuffle(self.questions_and_answers)


    def ask_question(self, question, correct_answer):
        user_answer = self.user_entry.get().strip().lower()
        if user_answer == correct_answer.lower():
            messagebox.showinfo("Correct!", "Correct answer!")
            return True
        else:
            messagebox.showerror("Wrong!", f"Wrong answer! The correct answer is: {correct_answer}")
            return False



    def ask_question(self, question, correct_answer):
        user_answer = self.user_entry.get().strip().lower()
        if user_answer == correct_answer.lower():
            messagebox.showinfo("Correct!", "Correct answer!")
            return True
        else:
            messagebox.showerror("Wrong!", f"Wrong answer! The correct answer is: {correct_answer}")
            return False

    def next_question(self):
        if self.current_question < len(self.questions_and_answers):
            question, _ = self.questions_and_answers[self.current_question]
            self.set_question_text(question)
            self.current_question += 1
            self.user_entry.delete(0, tk.END)
        else:
            messagebox.showinfo("Game Over", f"Your final score: {self.score}/{len(self.questions_and_answers)}")
            self.root.destroy()

    def show_score(self):
        messagebox.showinfo("Score", f"Your final score: {self.score}/{len(self.questions_and_answers)}")

    def set_question_text(self, question):
        max_chars = 150
        truncated_question = question[:max_chars] + ('...' if len(question) > max_chars else '')
        self.question_label.config(text=f"Question {self.current_question + 1}: {truncated_question}")

    def change_language(self, new_language):
        self.language = new_language
        self.initialize_questions()
        self.current_question = 0
        self.next_question()

    def run_quiz(self):
        self.root = tk.Tk()

        self.root.title("Football Quiz Game")
        self.root.geometry("700x500")  # Adjusted size

        self.root.configure(bg="#3498db")

        self.quiz_type_label = tk.Label(self.root, text="Choose Quiz Type:", font=("Helvetica", 14), bg="#3498db", fg="#ffffff")
        self.quiz_type_label.pack(pady=10)

        self.premier_league_button = tk.Button(self.root, text="Premier League", command=lambda: self.start_quiz('premier_league'), font=("Helvetica", 12), bg="#2ecc71", fg="#ffffff")
        self.premier_league_button.pack(pady=5)

        self.coupe_du_monde_button = tk.Button(self.root, text="Coupe du Monde", command=lambda: self.start_quiz('coupe_du_monde'), font=("Helvetica", 12), bg="#e74c3c", fg="#ffffff")
        self.coupe_du_monde_button.pack(pady=5)

        self.question_label = tk.Label(self.root, text="", font=("Helvetica", 14), wraplength=600, bg="#3498db", fg="#ffffff")
        self.question_label.pack(pady=20)

        self.user_entry = tk.Entry(self.root, font=("Helvetica", 12))
        self.user_entry.pack(pady=5)

        self.submit_button = tk.Button(self.root, text="Submit Answer", command=self.handle_submission, font=("Helvetica", 12), bg="#2ecc71", fg="#ffffff")
        self.submit_button.pack(pady=5)

        self.next_question_button = tk.Button(self.root, text="Next Question", command=self.next_question, font=("Helvetica", 12), bg="#e74c3c", fg="#ffffff")
        self.next_question_button.pack(pady=5)

        self.show_score_button = tk.Button(self.root, text="Show Score", command=self.show_score, font=("Helvetica", 12), bg="#3498db", fg="#ffffff")
        self.show_score_button.pack(pady=5)

        self.language_button = tk.Button(self.root, text="Change Language", command=self.change_language_prompt, font=("Helvetica", 12), bg="#3498db", fg="#ffffff")
        self.language_button.pack(pady=5)

        self.root.mainloop()

    def start_quiz(self, quiz_type):
        self.score = 0
        self.current_question = 0

        if quiz_type == 'premier_league':
            self.questions_and_answers = list(zip(self.questions_premier_league, self.answers_premier_league))
        elif quiz_type == 'coupe_du_monde':
            self.questions_and_answers = list(zip(self.questions_coupe_du_monde, self.answers_coupe_du_monde))
        else:
            messagebox.showerror("Error", "Invalid quiz type selected.")
            return

        random.shuffle(self.questions_and_answers)

        self.next_question()

    def handle_submission(self):
        if self.current_question > 0:
            question, correct_answer = self.questions_and_answers[self.current_question - 1]
            if self.ask_question(question, correct_answer):
                self.score += 1

    def change_language_prompt(self):
        if tk.TkVersion >= 8.6:
            lang_choice = simpledialog.askstring("Choose Language", "Enter language code ( fr:french, en:english, ar:arabic):")
        else:
            lang_choice = None

        if lang_choice:
            self.change_language(lang_choice.lower())

# Example usage
quiz_game = FootballQuizGame()
quiz_game.run_quiz()
```
